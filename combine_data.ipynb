```python
# hospital_df.groupby('Country Name')['Country Name'].count()
# hospital_df[].value_counts()

hospital_df = pd.read_csv('data/Hospital_Ranking_Data_2023.csv')
hospital_df=hospital_df.loc[hospital_df['Global Ranking']!='N']

hospital_df['Global Ranking'] = pd.to_numeric(hospital_df['Global Ranking'])
hospital_df['Hospital Size'] = pd.to_numeric(hospital_df['Hospital Size'])

grouped_hosp = pd.concat([hospital_df.groupby('Country Name')['Country Name'].count(),hospital_df.groupby('Country Name')[['Global Ranking','Hospital Size']].agg(['min','max'])],axis='columns')
# grouped_hosp.columns = ['Number of Ranked Hospitals', 'Average Ranking', 'Best Ranking', 'Lowest Ranking', 'Average Hospital Size','Smallest Hospital', 'Largest Hospital']
grouped_hosp.columns = ['Number of Ranked Hospitals','Best Ranking', 'Lowest Ranking', 'Smallest Hospital', 'Largest Hospital']
df = pd.merge(grouped_hosp,health_info[['Population','Life expectancy','Physicians per thousand', 'Infant mortality','Maternal mortality ratio']],left_index=True,right_index=True).dropna()
df['Population']=pd.to_numeric(df['Population'].str.replace(',',''))
df['Estimated Number of Physicians']=df['Population']/1000 * df['Physicians per thousand']
df['Normalized Population'] =df['Population']/(df['Population'].mean())
df['Normalized Life Expectancy'] =(df['Life expectancy']/df['Life expectancy'].mean())*10
df['Normalized Infant Mortality'] =df['Infant mortality']/df['Infant mortality'].mean()
df['Normalized Maternal Mortality'] =df['Maternal mortality ratio']/df['Maternal mortality ratio'].mean()
df['Normalized Ranked Hospital Count'] =df['Number of Ranked Hospitals']/df['Number of Ranked Hospitals'].mean()
df['Ranked Hospitals per Million'] = df['Number of Ranked Hospitals']/(df['Population']/1000000)
df['Normalized Hospitals per Population'] = (df['Number of Ranked Hospitals']/df['Population'])/(df['Number of Ranked Hospitals']/df['Population']).mean()
df = df.join(happiness[['Country','Ladder score',]].set_index('Country')).join(prosperity[['Country','AveragScore','Health']].set_index('Country'))
df
```
